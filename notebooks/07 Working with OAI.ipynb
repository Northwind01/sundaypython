````python
from sickle import Sickle
````

````python
sickle = Sickle('http://digital.slub-dresden.de/oai/')
````

````python
formats = sickle.ListMetadataFormats()
for f in formats:
    print(f.metadataPrefix)
````

````python
sets = sickle.ListSets()
for s in sets:
    print("%s (%s)" % (s.setName, s.setSpec))
````

````python
records = sickle.ListRecords(metadataPrefix='oai_dc', set="the-bicycle")
````

````python
record = records.next()
````

````python
record
````

````python
record.header.datestamp
````

````python
record.metadata.items()
````

````python
print(json.dumps(record.metadata, indent=4))
````
