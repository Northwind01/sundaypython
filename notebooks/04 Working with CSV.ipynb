```python
import pandas as pd
```

```python
url = "https://speicherwolke.uni-leipzig.de/index.php/s/2wjvzYDJyZWdzVm/download?path=%2F&files=HMT1-6200csv.csv"
```

```python
df = pd.read_csv(url, sep=";", encoding="latin-1")
```

```python
df
```

```python
df.info()
```

```python
df[df['JahrInskription'].isnull()]
```

```python
df = df.dropna().copy()
```

```python
df["JahrInskription"] = df["JahrInskription"].astype(int)
```

```python
df
```

```python
# Beliebte Vornamen
```
